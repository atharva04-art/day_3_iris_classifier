```python
import pandas as pd 
from sklearn.datasets import load_iris

# Load the Iris dataset from scikit-learn
data = load_iris()

#  Convert it into a DataFrame for easier handling 
df = pd.DataFrame(data.data, columns=data.feature_names)
df["target"] = data.target

df.head()
```

```python
from sklearn.model_selection import train_test_split

#  Features = all columns excep
x = df.drop("target", axis=1)

#  Lables = target column
y = df["target"]

#  Split data:80% for training, 20% for testing
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)
```

```python
print(df.columns)
```

```python
from sklearn.linear_model import LogisticRegression

# Create the model
model = LogisticRegression(max_iter=200)

# Train the model using training data
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

#  Make prediction on test data
y_pred = model.predict(x_test)

#  Check how accurate the model is
acc = accuracy_score(y_test, y_pred)

print("Model Accuracy:", acc)
```

```python
import pandas as pd
import numpy as np

# Use the same column names as original DataFrame
sample_df = pd.DataFrame([[6.1, 2.9, 4.7, 1.4]], columns=df.columns[:-1])

prediction = model.predict(sample_df)
print("Predicted Class:", data.target_names[prediction[0]])
```
